#Medias dos tempos

```python
mediaT_c_500v = np.mean(tempo_c_500v)
mediaT_c_2500v = np.mean(tempo_c_2500v)
mediaT_c_5000v = np.mean(tempo_c_5000v)

mediaT_py_500v = np.mean(tempo_py_500v)
mediaT_py_2500v = np.mean(tempo_py_2500v)
mediaT_py_5000v = np.mean(tempo_py_5000v)
```

```python
df_medias = pd.DataFrame({
    'Vertices': ['500', '2500', '5000'] * 2,
    'Tempo Médio': [mediaT_py_500v, mediaT_py_2500v, mediaT_py_5000v,
                    mediaT_c_500v, mediaT_c_2500v, mediaT_c_5000v],
    'Linguagem': ['Python'] * 3 + ['C'] * 3
})

# Plotagem
sns.set(style="whitegrid")
plt.figure(figsize=(8, 6))
sns.barplot(data=df_medias, x='Vertices', y='Tempo Médio', hue='Linguagem')

plt.title('Tempo Médio de Execução - Dijkstra (30 execuções)')
plt.ylabel('Tempo Médio (s)')
plt.xlabel('Quantidade de Vértices')
plt.tight_layout()
plt.show()

print("[ C ]")
print(f"Media dos Tempos das 30 execuções do algoritmo de Dijkstra; 500 Vertices: {mediaT_c_500v:.6f}")
print(f"Media dos Tempos das 30 execuções do algoritmo de Dijkstra; 2500 Vertices: {mediaT_c_2500v:.6f}")
print(f"Media dos Tempos das 30 execuções do algoritmo de Dijkstra; 5000 Vertices: {mediaT_c_5000v:.6f}")

print()

print("[ Python ]")
print(f"Media dos Tempos das 30 execuções do algoritmo de Dijkstra; 500 Vertices: {mediaT_py_500v:.6f}")
print(f"Media dos Tempos das 30 execuções do algoritmo de Dijkstra; 2500 Vertices: {mediaT_py_2500v:.6f}")
print(f"Media dos Tempos das 30 execuções do algoritmo de Dijkstra; 5000 Vertices: {mediaT_py_5000v:.6f}")
```

#Diferença Percentual e Multiplicidade de Velocidade

```python
vezes_mais_rapido_500v = mediaT_py_500v / mediaT_c_500v
vezes_mais_rapido_2500v = mediaT_py_2500v / mediaT_c_2500v
vezes_mais_rapido_5000v = mediaT_py_5000v / mediaT_c_5000v

vezes_velocidade = [vezes_mais_rapido_500v, vezes_mais_rapido_2500v, vezes_mais_rapido_5000v]
media_velocidade = np.mean(vezes_velocidade)

print(f"C foi {vezes_mais_rapido_500v:.2f}x mais rápido que Python (500v)")
print(f"C foi {vezes_mais_rapido_2500v:.2f}x mais rápido que Python (2500v)")
print(f"C foi {vezes_mais_rapido_5000v:.2f}x mais rápido que Python (5000v)")

print()

tempo_gasto_perc_500v = (mediaT_c_500v / mediaT_py_500v) * 100
tempo_gasto_perc_2500v = (mediaT_c_2500v / mediaT_py_2500v) * 100
tempo_gasto_perc_5000v = (mediaT_c_5000v / mediaT_py_5000v) * 100

tempos_gastos = [tempo_gasto_perc_500v, tempo_gasto_perc_2500v, tempo_gasto_perc_5000v]
media_tempo_gasto = np.mean(tempos_gastos)

print(f"C usou {tempo_gasto_perc_500v:.2f}% do tempo do Python (500v)")
print(f"C usou {tempo_gasto_perc_2500v:.2f}% do tempo do Python (2500v)")
print(f"C usou {tempo_gasto_perc_5000v:.2f}% do tempo do Python (5000v)")

print()

print(f"Conclusão:")
print(f"C é em média {media_velocidade:.2f}x mais rápido que Python para esse tipo de Algoritmo")
print(f"C gasta em média {media_tempo_gasto:.2f}% do tempo que Python gastaria para esse tipo de Algoritmo")
```

#Plotagem dos Graficos em linha para Comparação

```python

df = pd.DataFrame({
    "Python_500": tempo_py_500v,
    "C_500": tempo_c_500v,
    "Python_2500": tempo_py_2500v,
    "C_2500": tempo_c_2500v,
    "Python_5000": tempo_py_5000v,
    "C_5000": tempo_c_5000v
})


df_500 = df[['Python_500', 'C_500']].reset_index().melt(
    id_vars='index',
    value_vars=['Python_500', 'C_500'],
    var_name='Linguagem',
    value_name='Tempo'
)
df_2500 = df[['Python_2500', 'C_2500']].reset_index().melt(
    id_vars='index',
    value_vars=['Python_2500', 'C_2500'],
    var_name='Linguagem',
    value_name='Tempo'
)
df_5000 = df[['Python_5000', 'C_5000']].reset_index().melt(
    id_vars='index',
    value_vars=['Python_5000', 'C_5000'],
    var_name='Linguagem',
    value_name='Tempo'
)
```

```python
# Plotagens
sns.set(style="whitegrid")
plt.figure(figsize=(14, 8))

# Plot 500
plt.subplot(3, 1, 1)
sns.lineplot(data=df_500, x='index', y='Tempo', hue='Linguagem', marker='o')
plt.title("Desempenho para 500 Vértices")
plt.xlabel("")
plt.ylabel("Tempo (s)")

# Plot 2500
plt.subplot(3, 1, 2)
sns.lineplot(data=df_2500, x='index', y='Tempo', hue='Linguagem', marker='o')
plt.title("Desempenho para 2500 Vértices")
plt.xlabel("")
plt.ylabel("Tempo (s)")

# Plot 5000
plt.subplot(3, 1, 3)
sns.lineplot(data=df_5000, x='index', y='Tempo', hue='Linguagem', marker='o')
plt.title("Desempenho para 5000 Vértices")
plt.xlabel("Execução")
plt.ylabel("Tempo (s)")

plt.tight_layout()
plt.show()
```
